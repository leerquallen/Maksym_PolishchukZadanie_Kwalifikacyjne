```python
import pandas as pd
```

```python
data = pd.read_csv("data.csv", sep='\t')
prices = pd.read_csv("prices.csv", sep='\t')
quantity = pd.read_csv("quantity.csv", sep='\t', names=["part_number", "quantity"])
quantity.columns = ["part_number", "quantity"]
```

```python
merged_data = data.merge(prices, on='part_number', how='outer')
merged_data = merged_data.merge(quantity, on='part_number', how='outer')
```

```python
merged_data = merged_data[(merged_data['price'] != "0")]
merged_data = merged_data.dropna(subset=['quantity'])
```

```python
result = merged_data[['part_number', 'manufacturer', 'price', 'quantity']]
```

```python
result.head()
```

```python
brand_counts = result['manufacturer'].value_counts()
for brand, count in brand_counts.items():
    print(f'{brand} — {count} rows')
```

```python
result.to_csv('task_1_no_aggregate_quantities.csv', index=False, sep='\t')
```
